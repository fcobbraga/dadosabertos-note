```python
import requests
import pandas as pd
```

```python
url = 'https://dadosabertos.camara.leg.br/api/v2/proposicoes?ordem=ASC&ordenarPor=id'
response = requests.get(url)
```

```python
df_proposicao = pd.json_normalize(response.json()['dados'])
df_proposicao.head()
```

```python
df_proposicao.dtypes
```

```python
dict_proposicao_id = {}
for index, row in df_proposicao.iterrows():
    id_proposicao = row['id']
    dict_proposicao_id[id_proposicao] = {
        'sigla_tipo': row['siglaTipo'],
        'numero': row['numero'],
        'ano': row['ano'],
        'ementa': row['ementa']
    }
```

```python
dict_proposicao_id[16802]
```

```python
url_base = 'https://dadosabertos.camara.leg.br/api/v2/proposicoes/'
dados_proposicoes_id = []

#verificar se o dicionario (abaixo) esta correto para o webservice em questao
for id_proposicao in dict_proposicao_id.keys():
    url_proposicao = url_base + str(id_proposicao)
    response = requests.get(url_proposicao)
    json_proposicao = response.json()
    df_proposicao_id = pd.json_normalize(json_proposicao)
    dados_proposicoes_id.append(df_proposicao_id)

df_proposicoes_id = pd.concat(dados_proposicoes_id)
```
